# O problema

Ao medir a tensão de saída do sensor de silício com o multímetro numa certa configuração, obtivemos $\sim5\text{ V}$. Ao repetirmos a medida nesta mesma configuração, desta vez com o osciloscópio, obtivemos $\sim1\text{ V}$. Finalmente, medindo com o _lock-in_, obtivemos $\sim50\text{ mV}$.

# Hipótese

O comportamento observado sugere que a impedância dos equipamentos de medida estão interferindo no circuito do sensor de silício. Após uma série de medidas exploratórias e a consulta ao manual do sensor, chegamos à conclusão de que ele atua como fonte de corrente, não de tensão. Isto é, ele fornece uma corrente específica para uma dada potência luminosa incidente (segundo o manual, até $2.5\text{mA}/\text{W}$).

Contudo, isso não basta para explicar o que observamos, pois se é verdade que a corrente é fixada, então deveríamos observar uma queda de tensão maior no _lock-in_, seguida do osciloscópio e, por fim, o multímetro. Isso porque esta é a ordem decrescente de impedência desses equipamentos. Na verdade, o que devemos evidenciar é que, ao aumentar a impedância colocada *paralelamente* ao circuito do sensor, a corrente fornecida deve reduzir (isto é, a fonte "arriou").

# Metodologia

Para verificar a hipótese acima, utilizamos este circuito:

```python
Image(filename='../images/Lock-in_V.png')
```

Deste modo, a carga máxima colocada na saída do sensor de silício pode ser controlada pela resistência $R$: como ela está ligada em paralelo com o _lock-in_, que tem resistência interna muito maior, a resistência equivalente é aproximadamente igual a $R$.

Assim, a metodologia consiste em **medir a tensão** sobre o resistor $R$ e calcular a corrente, para valores de $R$ variando de $51\text{ k}\Omega$ a $11.6 \text{ M}\Omega$. Esperamos que, para algum intervalo de $R$, a corrente calculada mantenha-se constante (proporcional apenas à potência luminosa incidente). Finalmente, podemos **medir a corrente** diretamente com o _lock-in_ e compará-la com a corrente calculada. Esperamos que essa corrente seja igual àquela que se mantém aproximadamente constante.

# Análise

Abaixo estão as configurações desta análise: <code>PATH</code> é o caminho para os arquivos das medidas efetuas, organizadas num vetor de duplas (<code>MEASUREMENTS</code>): o primeiro elemento é o nome do arquivo da medida e o segundo, o valor da resistência colocada como carga do circuito do sensor de silício. <code>SCALE</code> é a escala usada para armazenar as medidas de tensão: mV (ou seja, os valores armazenados nos arquivos são todos dados em mV). <code>HEADER</code> indica a quantidade de linhas do cabeçalho do arquivo de medida, isto é, a quantidade de linhas que serão ignoradas:

```python
PATH = "../measurements" # Caminho da pasta que contém
MEASUREMENTS = [
  # (nome do arquivo, carga em mega-ohms)
    ("19.med",  0.051e6),
    ("33.med",  0.127e6),
    ("31.med",  0.500e6),
    ("28.med",  0.990e6),
    ("30.med",  2.000e6),
    ("29.med",  3.330e6),    
    ("27.med",  4.800e6),
    ("23.med",  6.800e6),
    ("26.med",  7.800e6),
    ("25.med", 11.600e6)
]
SCALE = 1e-3 # As medidas do lock-in são gravadas em mV
HEADER = 5 # Quantidade de linhas de cabeçalho
```

Abaixo, a primeira linha carrega as várias medidas declaradas acima num vetor de <code>DataSet</code>. Cada elemento do vetor representa uma medida: comprimento de onda (em nanometros) e tensão medida (em milivolts).

O loop no final transforma a medida de tensão em corrente, com base no resistor colocado como carga do sensor de silício (a medida de tensão do lock-in foi feita sobre esse resistor), conforme associação declarada acima.

```python
data = [pd.read_csv(PATH + "/" + filename, sep="\t", header=HEADER) for (filename,_) in MEASUREMENTS]

# TODO: como evitar o "for"?
for i in range(len(data)):
    r = MEASUREMENTS[i][1]
    data[i]["Magnitude (mV)"] = np.multiply(data[i]["Magnitude (mV)"], SCALE / r) / 1e-9
```

Desenha os gráficos das medidas.

```python
# TODO: como evitar o "for"?
for i in range(len(data)):
    plt.plot(data[i]["Comprimento de onda(nm)"], data[i]["Magnitude (mV)"],
             label=str(MEASUREMENTS[i][0] + ": " + str(MEASUREMENTS[i][1]/1e6) + " M$\Omega$"))
    
FONT = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 16,
        }
plt.xlabel('$\lambda$ (nm)', fontdict=FONT)
plt.ylabel('$I$ (nA)', fontdict=FONT)
plt.legend()
```

Acho que mais fácil que analisar o gráfico acima é observar a corrente máxima calculada, que ocorre no comprimento de onda 980 nm:

```python
x = []
y = []
# TODO: como evitar o "for"?
for i in range(len(data)):
    x.append(MEASUREMENTS[i][1]/1e6)
    y.append(max(data[i]["Magnitude (mV)"]))    

plt.plot(x, y, 'ro-')
plt.xlabel('$R$ (M$\Omega$)', fontdict=FONT)
plt.ylabel('$I$ (nA)', fontdict=FONT)
plt.show()
```

Note como para $R \le 2\text{ M}\Omega$, a corrente _calculada_ é aproximadamente constante: 36-41 nA. Mas se a carga for superior a esse valor, a corrente fornecida pelo circuito do sensor de silício cai.

Para comparação, abaixo está a medida de **corrente** efetuada pelo lock-in, nas mesmas condições das medidas acima (comente/descomente a linha indicada abaixo para ocultar/exibir também o gráfico da corrente _calculada_ para o caso da carga mais baixa).

```python
FILE = "35.med" # Arquivo da medida de corrente
data_I = pd.read_csv(PATH + "/" + FILE, sep="\t", header=HEADER) # Carrega o arquivo
data_I["Magnitude (mV)"] = data_I["Magnitude (mV)"] * 1e6 # Corrente em nA
plt.plot(data_I["Comprimento de onda(nm)"], data_I["Magnitude (mV)"], label=str(FILE)) # Desenha o gráfico
#plt.plot(data[0]["Comprimento de onda(nm)"], data[0]["Magnitude (mV)"], label="19.med") # <---- COMENTE/DESCOMENTE
plt.xlabel('$\lambda$ (nm)', fontdict=FONT)
plt.ylabel('$I$ (nA)', fontdict=FONT)
plt.legend()
plt.show()
```
